```python
import fastf1 as ff1
import pandas as pd

# Tutorial para scripts rápidos e dinâmicos para análise de dados de F1

#Habilitando cache

ff1.Cache.enable_cache('cache/')

# Carregando a temporada e corrida

race = ff1.get_session(2022, 'Belgian', 'R')

# Carregando as voltas

laps = race.load_laps(with_telemetry=True)


```

```python
# Exporando dados. Panda fornece as funções head() e tail() para auxiliar

laps.head()

```

```python
laps.tail()
```

```python
# Usando a função 'T' para coletar mais dados

laps.T
```

```python
laps.head().T
```

```python
# Descobrindo a melhor volta com DataFrame

laps.loc[laps['LapTime'] == min(laps['LapTime'].dropna())]
```

```python
# Contando a usagem total de pneus de todo o pelotão, na corrida

laps['Compound'].value_counts()
```

```python
# Consulta rápida para ritmo de corrida, usando groupby de Pandas

laps[['Team', 'Driver', 'LapTime']].groupby(['Team', 'Driver']).mean().sort_values(['Team', 'LapTime'])

```

```python
# Coletando as maiores velocidades alcançadas na corrida

laps[['Driver', 'SpeedST']].groupby(['Driver']).max().sort_values('SpeedST', ascending=False)
```
