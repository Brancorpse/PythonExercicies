# Relacionando variávais com scatterplots


```python
# verificando em gráfico a quantidade de vacinados e a de totalmente vacinados

sns.relplot(x="people_fully_vaccinated", y= "people_fully_vaccinated", data=data)
```

```python
# verificando total de casos com o total de mortes em gráfico 

sns.relplot(x="total_cases", y= "total_deaths", data=data)
```

```python
# relação entre total de casos e novos casos

sns.relplot(x="total_cases", y= "new_cases", data=data)
```

```python
# relação entre total de hospitalizações semanais x total de entradas na uti

sns.relplot(x="weekly_hosp_admissions", y= "weekly_icu_admissions", data=data)
```
